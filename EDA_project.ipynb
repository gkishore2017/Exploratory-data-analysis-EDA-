# Boxplots : is for outliers detection


    1. Boxplot / WHISKERPLOT 
    2. Scatter plot
    3. Z Score/ Standard Deviation method (Data points which are behond + - 3 standard deviation are treated as outliers )
    4. IQR Method (inter quartile range) : 

```python
sns.boxplot(df['bmi'])
```

```python
q3 = df['bmi'].quantile(.75)
q1 = df['bmi'].quantile(.25)

print('Q3 :',q3)
print('Q1 :',q1)
```

```python
IQR = q3 - q1
print('IQR :',IQR)
```

```python
Min_WHISKERPLOT = q1 -(1.5 * IQR )
Max_WHISKERPLOT = q3 +(1.5 * IQR)

print('Max_WHISKERPLOT :',Max_WHISKERPLOT)
print('Min_WHISKERPLOT :',Min_WHISKERPLOT)
```
